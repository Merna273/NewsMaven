```python
# select randomly 1 file from each folder in each folder in CleanData
def selectRandomFile():
    folders = []
    foldersPaths=[]
    subfolders=["AA","AB","AC","AD","AE"]

    for i in range(0,40):
        folders.append("wiki_pages_" + str(i))
    for folder in folders:
        for subfolder in subfolders:
            path = "CleanData/" + folder + "/" + subfolder
            files = os.listdir(path)
            random_file = random.choice(files)
            foldersPaths.append(path + "/" + random_file)
    return foldersPaths
```

```python
filesPaths=selectRandomFile()
for path in filesPaths:
    print(path)
```

```python
def load_json_lines(file_path):
    """Load JSON objects from a file that contains one JSON object per line."""
    articles = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()  # Remove leading/trailing whitespace
            if line:  # Make sure the line isn't empty
                try:
                    articles.append(json.loads(line))
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON on line: {line}\nError: {e}")
    return articles

def get_random_article(file_path):
    """Select a random article with non-empty text from the JSON file."""
    articles = load_json_lines(file_path)
    
    # Filter out articles with empty "text" field
    non_empty_articles = [article for article in articles if article.get("text")]

    if non_empty_articles:
        return random.choice(non_empty_articles)
    else:
        raise ValueError("No articles with non-empty text found.")

# Load a random article with non-empty text from filesPaths[0]
random_article = get_random_article(filesPaths[0])

# Print the random article
print(json.dumps(random_article, indent=4))
```

```python
# load a random article from each file in filesPaths, then write that article to a json file
def write_random_articles_to_files(file_paths):
    counter=0
    for file_path in file_paths:
        random_article = get_random_article(file_path)
        # write the random_article to a json file titled random_article_0
        with open("random_article_"+str(counter)+".txt", 'w', encoding='utf-8') as file:
            file.write(json.dumps(random_article["title"], indent=4))
            file.write("\n")
            file.write(json.dumps(random_article["text"], indent=4))


        counter+=1

write_random_articles_to_files(filesPaths)
 
```
